```
from IPython.display import clear_output
from math import log2, ceil, floor
import textwrap

class Binary():
  def __init__(self, val, length, exp_length):
    val = "".join(['0' for n in range(0, length)]) if val == None else val
    assert len(val) == length, "Não possui " + str(length) + " bits"
    
    self.exp_bias = 2 ** (self.exp_length - 1) - 1
    self.digit    = "".join([self.assert_binary(n) for n in list(val)[0]])
    self.expoent  = "".join([self.assert_binary(n) for n in list(val)[1 : exp_length + 1]])
    self.mantissa = "".join([self.assert_binary(n) for n in list(val)[exp_length + 1:]])
 
  def assert_binary(self, n):
    assert n in ('0', '1'), "Não é binário"
    return n
  
  def convert_int_to_bin(self, val):
    result, val = '', int(val)
    while val > 1:
      result = str(val % 2) + result
      val = int(val / 2)
    return str(val) + result
 
  def convert_dec_to_bin(self, val):
    result = ''
    for i in range(0, self.mant_length + 2):
      val = float(val) * 2
      result += f'{val:.60f}'[0]
      val = '0' + f'{val:.60f}'[1:]
    return result

  def convert_bin_to_int(self, val):
    result = 0
    for i in range(len(val)):
      result += int(val[i]) * 2 ** (len(val) - 1 - i)    
    return str(result)
 
  def get_binary(self):
    return self.digit + self.expoent + self.mantissa

  def get_hex(self):
    bin_val = self.get_binary()
    hex_val = ''
    bin_splits = textwrap.wrap(bin_val, 4)
    for bin_val in bin_splits:
      hex_val += hex(int(self.convert_bin_to_int(bin_val)))[2:]
    return hex_val

  def round_mantissa(self):
    tmp = self.mantissa[self.mant_length]
    self.mantissa = list(self.mantissa[:-1])
    for i in range(self.mant_length - 1, 0, -1):
      self.mantissa[i] = str(int(self.mantissa[i]) + int(tmp))
      if self.mantissa[i] == '2':
        self.mantissa[i] = '1'
      else:
        break
    self.mantissa = "".join([n for n in self.mantissa])
    return 1 if i == 0 else 0

  def decimal_to_binary(self, val):
    int_part, dec_part = val.split('.')
    dec_part = '0.' + dec_part
    if int_part[0] == '-':
      self.digit = '1'
      int_part = int_part[1:]
    else:
      self.digit = '0'

    if int_part == '0':
     exp = floor(log2(float(dec_part)))
     if -exp >= self.exp_bias:
      exp = -self.exp_bias
      dec_part = float(dec_part) / 2
      
     dec_part = f'{(float(dec_part) / 2 ** (exp + 1)):.60f}'
     tmp_mant = self.convert_dec_to_bin(dec_part)[1:]
    else:
      exp = floor(log2(float(int_part)))
      tmp_mant = self.convert_int_to_bin(int_part)[1:] + self.convert_dec_to_bin(dec_part)

    self.mantissa = tmp_mant + "".join(['0' for n in range(self.mant_length - len(tmp_mant) + 1)]) if len(tmp_mant) < self.mant_length else tmp_mant[:self.mant_length + 1]
    excess_bit = self.round_mantissa()
    
    exp_val = exp + self.exp_bias + excess_bit
    tmp_exp = self.convert_int_to_bin(exp_val)
    self.expoent = "".join(['0' for n in range(self.exp_length - len(tmp_exp))]) + tmp_exp if len(tmp_exp) < self.exp_length else tmp_exp

class _8bit_(Binary):
  length, exp_length, mant_length = 8, 3, 4 
  def __init__(self, val = None):
    super().__init__(val, self.length, self.exp_length)

class _32bit_(Binary):
  length, exp_length, mant_length = 32, 8, 23 
  def __init__(self, val = None):
    super().__init__(val, self.length, self.exp_length)

class _64bit_(Binary):
  length, exp_length, mant_length = 64, 11, 52 
  def __init__(self, val = None):
    super().__init__(val, self.length, self.exp_length)
```

```
while True:
  clear_output()
  print("A) Converter número decimal para binário 8 bits:")
  print("B) Converter número decimal para binário 32 bits:")
  print("C) Converter número decimal para binário 64 bits:")
  print("D) Sair do programa:")
  char = input("")
  
  if char in ('A', 'a'):
    x = _8bit_()
  elif char in ('B', 'b'):
    x = _32bit_()
  elif char in ('C', 'c'):
    x = _64bit_()
  else:
    break
  
  clear_output()
  num = input("Digite o valor na base 10 a ser convertido: ")
  x.decimal_to_binary(f'{float(num):.60f}')
  clear_output()
 
  while True:
    print("Resultado da conversão de " + num + " -> " + x.get_binary())
    print("A) Exibir dígito:")
    print("B) Exibir Expoente:")
    print("C) Exibir Mantissa:")
    print("D) Converter em Hexadecimal:")
    print("E) Voltar ao menu inicial:")
    char = input("")
    clear_output()
 
    if char in ('A', 'a'):
      print("Dígito: " + x.digit)
    elif char in ('B', 'b'):
      print("Expoente: " + x.expoent)
    elif char in ('C', 'c'):
      print("Mantissa: " + x.mantissa)
    elif char in ('D', 'd'):
      print("Valor em Hexadecimal: " + x.get_hex())
    elif char in ('E', 'e'):
      break
```
